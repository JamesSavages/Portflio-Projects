```python
# import libraries
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Now we need to read in the data
df = pd.read_csv(r"C:\Users\james\OneDrive\Desktop\Data Analysis\Python\Dataset\movies.csv")
```

```python
# let view the data
df.head()
```

```python
# Lets see if there is missing data
for col in  df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
# Review the data types of our columns
df.dtypes
```

```python
# create correct year column
df['year_correct'] = df['released'].astype(str).str[:4]
```

```python
# Order data by gross revenwu
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# we will now test the correlation between gross & other fields
```

```python
# Scatter plot with buget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget Vs Gross Earning')
plt.xlabel('Budget of Film')
plt.ylabel('Gross Earnings')
plt.show
```

```python
# Lets look at some of the correlations
df.corr(method='pearson') # Pearsons, kendall & spearman are the different types. Default is pearsons
```

```python
#high correlation between budget & gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show

plt.show()
```

```python
# Look at company
df.head()
```

```python
# We will convert all fields to numeric
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
        
df_numeric.head()
```

```python
correlation_matrix = df_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show

plt.show()
```

```python
df_numeric.corr()
```

```python
correlation_mat = df_numeric.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5)]
high_corr
```
